```
install.packages("keras")
```

```
library(tidymodels)

mod_ref <- parsnip::linear_reg() #tipo de modelo que usaremos logístico. Aquí podéis probar otras opciones

model_log <- 
 mod_ref %>%  # 1. ¿Cuál es el algoritmo que usaremos? 
  set_engine("lm") %>% # 2. Que paquete usaremos para implementar el algoritmo
  fit_xy(
      x=train_x,
      y=train_y
      # 3. Nuestros datos
  )

model_log
```

```
predicciones = predict(model_log,test_x)
summary(predicciones)
```
