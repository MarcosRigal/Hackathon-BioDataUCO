```
my_symbol = c()
```

```
my_go = c()
```

```
myfiles = list.files()
```

Lo hemos limpiado porque 

```
for(i in seq(length(myfiles)))
{
  foo = read.delim(myfiles[i])
  my_symbol= c(my_symbol, foo$SYMBOL)
  my_go = c(my_go, foo$GO.TERM)
}

```

```
my_symbol
```

```
my_go
```

```
data <- data.frame(cbind(my_symbol, my_go))
```

```
data
```

```
install.packages("dplyr")
```

```
library(dplyr)
count(data)
```

```
library(dplyr)
count(distinct(data))
```

```
databien <- distinct(data)
```

```
databien
```

```
write.csv(databien,file = "/home/data/tSpm.csv")
```
